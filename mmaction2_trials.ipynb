```python
def download_video(video_id, download_path):
    import os
    # os.system(f"rm {input_video_path}")
    # downloads youtube video in mp4 format
    video_url = f'https://www.youtube.com/watch?v={video_id}'
    os.system(f"python -m youtube_dl -f 'bestvideo[ext=mp4]+bestaudio[ext=m4a]/mp4' {video_url} -o '{download_path}downloaded_video_{video_id}.%(ext)s'")
    # Print a success message
    print('Video downloaded successfully')
```
